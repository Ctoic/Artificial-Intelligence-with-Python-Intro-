## Online Store Shopping Cart
Write a Python program to simulate the online store's shopping cart system. The program should use a while loop to provide an interactive shopping experience.


### Tasks:
* Initialize an empty list called shopping_cart to store items.
* Display a welcome message and available products.
* Loop should be iterating over AI Response Simulator and Investment decision and keep asking for user input.
* Prompt the user to enter the number corresponding to the product they want to add to the cart.
* Check if the user entered 'q'. If so, break the loop.
* Convert the choice to an integer and validate it within the valid range (1 to 3).
* Based on the user's choice, add the selected product to the shopping_cart list.
* Display a confirmation message indicating the added product.
* Continue the loop to allow the user to add more products or proceed to checkout.
* After the loop ends, display the contents of the shopping cart.
* Display a goodbye message.
* End the program.


```python
def display_available_products():
    print("Available products:")
    print("1. AI Response Simulator - $99")
    print("2. Investment Decision - $199")
    print("3. ChatGPT T-shirt - $25")

def main():
    shopping_cart = []
    
    print("Welcome to the Online Store!")
    display_available_products()
    
    while True:
        choice = input("Enter the number corresponding to the product you want to add to the cart (or 'q' to quit): ")
        
        if choice.lower() == 'q':
            break
        
        try:
            choice = int(choice)
            if choice < 1 or choice > 3:
                raise ValueError
        except ValueError:
            print("Please enter a valid option (1 to 3) or 'q' to quit.")
            continue
        
        if choice == 1:
            shopping_cart.append("AI Response Simulator")
        elif choice == 2:
            shopping_cart.append("Investment Decision")
        elif choice == 3:
            shopping_cart.append("ChatGPT T-shirt")
        
        print("Product added to cart.")
    
    print("\nItems in your shopping cart:")
    for item in shopping_cart:
        print("- " + item)
    
    print("\nThank you for shopping with us! Goodbye.")

if __name__ == "__main__":
    main()
```

## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
def classify_email(email):
    if "important" in email:
        return "Important"
    elif "promotion" in email:
        return "promotion"
    else:
        return "spam"

# now creating a list of email where there will be three emails
emails = [
    "Hello, we would like to notify you that your application has been approved. Please check your account for more details.",
    "Hi, we have a special promotion just for you! Get 50% off on your first purchase.",
    "Congratulations! You have won a free iPhone! Click here to claim your prize."
]

# Now we will pass this email to classify_email function and print the result
for email in emails:
    print(classify_email(email))
    
```

Use the dataset of Game,  perform following operations.

* Calculate the total global sales for all games in the dataset.
* Find the game with the highest global sales.
* Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).
* Find the top 5 most popular game genres based on global sales.
* Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).
* Display last 10 rows.
* Determine the top 3 gaming platforms with the highest global sales.
* Calculate the market share (percentage of global sales) for each genre.

```python
# read data set named data.csv
import pandas as pd
data = pd.read_csv('data.csv' , delimiter=',')
# # calculate the total Global_Sales for all games in the dataset
total_sales = data['Global_Sales'].sum()
print("Totla Global Sales : " , total_sales)

# Find the game with highest Global_Sales
highest_sales = data['Global_Sales'].max()
print("Higest Global Sales : ", highest_sales)

# * Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).
average_sales = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
print("Average Sales for each region : ", average_sales)

#  Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).

total_sales_by_genre = data.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

print("Total Sales by Genre in each region : ", total_sales_by_genre)

# * Display last 10 rows.
print(data.tail(10))

#* Determine the top 3 gaming platforms with the highest global sales.
top_3_platforms = data.groupby('Platform')['Global_Sales'].sum().nlargest(3)
print("Top 3 gaming platforms with the highest global sales : ", top_3_platforms)


#* Calculate the market share (percentage of global sales) for each genre.

market_share = (data.groupby('Genre')['Global_Sales'].sum() / total_sales) * 100
print("Market share for each genre : ", market_share)

#* Calculate the correlation coefficient between NA_Sales and EU_Sales. (.corr function)
correlation = data['NA_Sales'].corr(data['EU_Sales'])
print("Correlation coefficient between NA_Sales and EU_Sales : ", correlation)














```

```python
#  Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales) plot 
total_sales_by_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales by Genre in each region')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.show()
```

### Simple Reflex agent:
Design an reflex agent in Python to simulate room-cleaning behavior, now capable of moving both right and left within the environment. Your task is to implement this improved reflex agent and demonstrate its functionality in a simulated environment.

Your code should include:

A SimpleReflexAgent class with the following methos:

def perceive(self, location, dirty)

def clean(self)

def move_right(self)

def move_left(self)
    
    
A simulate_environment function to simulate the agent's actions in the environment. 

def simulate_environment(agent, environment):


Demonstrative functionality showcasing the agent's behavior in response to the cleanliness of locations and its ability to move within the environment. Write the code and provide a sample output demonstrating the agent's behavior in a predefined environment.

```python
class SimpleReflexAgent:
    def __init__(self):
        self.location = 0  # Initial location
        self.dirty = False  # Initial state of cleanliness

    def perceive(self, location, dirty):
        self.location = location
        self.dirty = dirty

    def clean(self):
        print("Cleaning the room at location", self.location)
        self.dirty = False

    def move_right(self):
        print("Moving right from location", self.location)
        self.location += 1

    def move_left(self):
        print("Moving left from location", self.location)
        self.location -= 1


def simulate_environment(agent, environment):
    print("Initial state of the environment:", environment)

    for i in range(len(environment)):
        agent.perceive(i, environment[i])
        
        if agent.dirty:
            agent.clean()
        
        if i != len(environment) - 1:  # Move only if not at the last location
            if agent.location == 0:
                agent.move_right()

# Sample environment where 1 represents dirty and 0 represents clean
environment = [1, 0, 1, 1, 0, 1, 0, 0]

# Create a SimpleReflexAgent instance
agent = SimpleReflexAgent()

# Simulate the environment
simulate_environment(agent, environment)
```
